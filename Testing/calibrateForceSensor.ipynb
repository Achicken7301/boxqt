```python
rFRS = []
mass = range(1000, 10000)
for m in mass:
    rFRS.append(271*1000*(pow(m , -0.69)))
```

```python
plt.ylim([0, 4*1000])
plt.plot(mass, rFRS)
plt.ylabel("Resistance(ohm)")
plt.xlabel("Added mass(g)")
```

```python
# r0 = [100, 1000, 4700, 10*1000, 50*10000]
r0 = range(400, 4000, 100)
# r0 = 1000
vcc = 3.3
Vmeas = []
i = 0

plt.ylim([0, vcc])
plt.ylabel("Vmeas(V)")
plt.xlabel("mass(g)")
for r in r0:
    for m in mass:
        Vmeas.append((r * vcc) / (r + (271000 * pow(m, -0.69))))

    plt.plot(mass, Vmeas)
    print(str(max(Vmeas) - min(Vmeas)) + " r0: " + str(r))
    i += 1
    Vmeas = []
```

### Convert ADC to Voltage

```python
def map(x, in_min, in_max, out_min, out_max):
  return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
```
