```python
def map( x,  in_min,  in_max,  out_min,  out_max):
  return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min;
```

```python
def vol2gram(vol):
     vcc = 3.3
     r0 = 1000.0     
     m = vcc / float(vol)
     m = m - 1
     m = r0 * m
     m = 271000.0 / m
     m = pow(m, 1 / 0.69)
     return m

def fsr(vol):
     vcc = 3.3
     r0 = 1000.0   
     rFSR = vcc / float(vol)
     rFSR = rFSR - 1
     rFSR = r0 * rFSR
     return rFSR
```

```python
ser.close()
ser = serial.Serial(port="COM6", baudrate=9600)
start = time.time()
while True:
    m = 0
    
    
    b = ser.readline()
    data_ser = b.decode().splitlines()
    vol = float(map(float(data_ser[0]), 0, 4095, 0, 3.3))
    if (vol == 0):
        vol = 0.0000001
    # print("analog: " + str(data_ser[0]) + " vol: " + str(vol))
    
    # vcc / vmeas
    m = vcc / vol
    m = m - 1
    m = r0 * m
    m = 271000.0 / m
    m = pow(m, 1 / 0.69)
    m /= 1000
    print("vol: " + str(vol) + " mass: " + str(m)+"kg")
    if (time.time() - start > 1):
        print("Close")
        ser.close()
        break

# print(m)
```
